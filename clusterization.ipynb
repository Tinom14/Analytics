```python
df = df.dropna()
df.info()
```

```python
df.describe()
```

```python
df = df[(df['Вес'] <= 150) & (df['Возраст'] <= 100)]
df.describe()
```

```python
for i in df['Проффесия'].unique():
    column_name = 'Профессия' + i
    df[column_name] = df['Проффесия'].isin([i]).astype(int)
df = df.drop(['Проффесия'], axis = 1 )   
df
```

```python
df['Высшее_образование'] = df['Высшее_образование'].map({'Да' : 1, 'Нет' : 0})
df['Пол'] = df['Пол'].map({'М' : 0, 'Ж' : 1})
df
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
X = df
eps_array = [0.01, 0.2, 0.6, 1, 2, 5, 10, 20, 100]
min_samples_array = [1, 2, 3, 5, 10, 15, 20, 25, 30, 50]
best_eps = eps_array[0]
best_min_samples = min_samples_array[0]
best_silhouette_score = -1.0
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[X.columns] = scaler.fit_transform(X)
```

```python
X
```

```python
for eps in eps_array:
    for min_samples in min_samples_array:
        clustering = DBSCAN(eps = eps, min_samples = min_samples).fit(X)
        df['Cluster_rez'] = clustering.labels_
        if len(df['Cluster_rez'].unique()) <= 1:
            continue
        else:
            temp_silhouette = silhouette_score(X, df['Cluster_rez'])  
            if temp_silhouette > best_silhouette_score:
                best_silhouette_score = temp_silhouette
                best_eps = eps
                best_min_samples = min_samples
print(best_eps, best_min_samples, best_silhouette_score)
```
